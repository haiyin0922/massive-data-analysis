# Mapper
## shingling
* 題目說 "you should use 3-shingles"
* 所以先把每三個word連成一組，變成一個shingle

```python
def shingling(line):
    L = []
    for i in range(len(line[1])-2):
        L.append((line[1][i] + line[1][i+1] + line[1][i+2]))
    
    return (line[0], L)
```

## hash_shingle
* 將前面產生的shingles用python內建function做hash
* 並計算總共有多少個shingle

```python
def hash_shingle(line):
    L = []
    for i in line[1]:
        L.append(hash(i))
        
    global N    
    N += len(L)
    
    return (line[0], L)
```

## min_hash
* 使用h(x)=((ax+b) mod p) mod N 來做min hash

```python
def min_hash(line):
    L = []
    for i in range(101):
        M = 34012
        for j in line[1]:
            h = ((a[i]*j + b[i]) %34019) %34012
            if h < M:
                M = h
        L.append(M)
        
    return (line[0], L)
```

## LSH
* 將 signature matrix M partition into 50 bands and 2 rows
* 然後至少有一個band中的兩個row到同一個bucket的，就把它們加入candidate pairs

```python
def LSH(line):
    L = []
    for i in range(100):
        for j in range(i+1, 101):
            for k in range(50):
                if line[i][1][2*k] == line[j][1][2*k] and line[i][1][2*k+1] == line[j][1][2*k+1]:
                    L.append((line[i][0], line[j][0]))
                    break
    
    return L
```

## calculate_sim
* 使用Jaccard Similarity的定義來計算similarity
* | C1交集C2 | / | C1聯集C2 |
* 然後要求前十名，所以將結果排序

```python
def calculate_sim(line):
    L = []
    for i in range(len(line[1])):
        C1 = set(line[0][line[1][i][0]-1][1])
        C2 = set(line[0][line[1][i][1]-1][1])
        sim = len(C1.intersection(C2)) / len(C1.union(C2))
        L.append((sim, (line[1][i][0], line[1][i][1])))
    
    L_sort = sorted(L, reverse = True)
    return L_sort
```

## thesame_key
* 用來將data換成相同的key，方便做group

```python
def thesame_key(line):
    return (-1, line)
```

## read input
* 這次的input是資料夾中的所有資料，所以使用wholeTextFiles來讀取
* 並且將article中每個字分開，再分別組合成shingles

```python
conf = SparkConf().set("spark.default.paralleism", 4).setAppName("LSH")
sc = SparkContext.getOrCreate(conf=conf)
data = sc.wholeTextFiles("./athletics").map(lambda x: (int(x[0][-7:-4]), x[1])).map(lambda x: (x[0], x[1].split()))
shingle = data.map(shingling)

print(sorted(shingle.collect()))
```

## hash shingle
* 投影片第19頁說 "Represent a document by the set of hash values of its k-shingles"
* 所以先hash剛剛產生的shingles
* N用來計算總共有多少個shingle

```python
N = sc.accumulator(0)
shingle_hash = shingle.map(hash_shingle)

shingle_hash_collect = sorted(shingle_hash.collect())
print(shingle_hash_collect)
print(N)
```

## random a, b
* 投影片第44頁中有寫 "hash fucntion = ((ax+b) mod p) mod N"
* a,b … random integers
* p … prime number (p > N)
* 然後作業又要求 "you need 100 different hash functions"
* 為了避免用到重複的hash function，所以這邊我用shuffle的方式來取random a, b

```python
a = []
b = []
for i in range(34011):
    a.append(i+1)
    b.append(i+1)
shuffle(a)
shuffle(b)

print(a)
print(b)
```

## minhash
* 用剛剛產生的100個hash function進行min hash

```python
shingle_minhash = shingle_hash.map(min_hash)

shingle_minhash_collect = sorted(shingle_minhash.collect())
print(shingle_minhash_collect)
```

## LSH
* 將做完minhash的資料加上相同的key，再把他們group起來方便使用
* 然後做LSH，找出candidate pairs

```python
minhash_key = shingle_minhash.map(thesame_key).groupByKey().mapValues(list).map(lambda x: x[1])
shingle_LSH = minhash_key.map(LSH)

shingle_LSH_collect = sorted(shingle_LSH.collect())
print(shingle_LSH_collect)
```

## combine data
* 把minhash後data跟LSH後的data combine在一起
* 方便做similarity的運算

```python
minhash_key = shingle_minhash.map(thesame_key).groupByKey().mapValues(list)
LSH_key = shingle_LSH.map(thesame_key)
shingle_candidate = minhash_key.join(LSH_key).reduceByKey(lambda x, y: x + y).map(lambda x: x[1])

print(shingle_candidate.collect())
```

## calculate similarity
* 將剛剛產生的data丟進function計算sim值

```python
similarity = shingle_candidate.map(calculate_sim)

rank = similarity.collect()
print(rank)
```

## result
* 印出作業要求的結果前十名 (index1, index2) : similarity

```python
for i in range(10):
    print(str(rank[0][i][1]) + ':' + str(rank[0][i][0]))
```
